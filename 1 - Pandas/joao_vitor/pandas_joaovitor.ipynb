```python
import pandas as pd
import matplotlib.pyplot as plt

lista = pd.read_csv('compras.csv')
produtos = lista.itemDescription.unique()
print(produtos)
#media de vendas de cada produto
qntmedia = lista.groupby('itemDescription').agg({"quantidade": "mean"})
print(qntmedia)
plt.plot(qntmedia)
plt.show()
```

```python
#venda total de cada dia
vendadia = lista.groupby('Date').agg({"price": "sum"})
plt.plot(vendadia)
plt.show()
```
